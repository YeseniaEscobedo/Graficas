```python
import matplotlib.pyplot as plt 
x=[1,2,3,4]
y=[1,4,16,25]

plt.plot(x,y)
plt.show

plt.plot(x,y, label='x^2')
plt.title('Simple Line Plt')
plt.xlabel('x Axis')
plt.xlabel('y Axis')
plt.show
```

```python
categorias = ['A','B','C']
values= [7,3,5]
plt.bar(categorias,values)
plt.title('Bar Chart')
plt.xlabel('Categorias')
plt.ylabel('Values')
plt.show
```
